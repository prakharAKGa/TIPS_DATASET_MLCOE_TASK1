```python
fig, axes =plt.subplots(2,3, figsize=(16,8))
plt.tight_layout(pad=2)


#day vs total bill


axes[0,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='smoker', data=tips, ax=axes[0,0]);

#Total bill vs total tip

axes[0,1].set_title('Total Bill vs. Total Tip')
sns.scatterplot(x=tips.total_bill, 
                y=tips.tip, 
                hue=tips.sex, 
                s=100, 
                ax=axes[0,1]);


#total bill and tip vs day

axes[0,2].plot(tips.day,tips.total_bill, 'X--b')
axes[0,2].plot(tips.day,tips.tip, 'o--r')
axes[0,2].set_xlabel('DAY')
axes[0,2].set_ylabel('BILL(Rs.)')
axes[0,2].legend(['TOTAL_BILL', 'TIP']);
axes[0,2].set_title('GRAPH OF TOTAL BILL AND TIP');



#total bill and day and size

sns.barplot(x='day', y='total_bill', hue='size', data=tips, ax=axes[1,0]);

#total tip and day and size

sns.barplot(x='day', y='tip', hue='size', data=tips, ax=axes[1,1]);


#tip and day and time

sns.barplot(x='day', y='tip', hue='time', data=tips, ax=axes[1,2]);
```
